## Read data from an existing file

```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:\\Users\\asus\\OneDrive\\Desktop\\Python\\auto-mpg.csv")
df.head
```

## Engineers at XYZ custom cars want to know how many cars are Fuel efficient.

## MPG > 29, Horsepower < 93.5, Weight < 2500

```python
df.loc[(df['mpg'] > 29) & (df['horsepower'] < 93.5) & (df['weight'] < 2500)]
```

## Engineers at XYZ custom cars want to know how many cars are Muscle cars.

## Displacement>262, Horsepower>126, Weight in range[2800 - 3600]

```python
df.loc[(df['displacement'] > 262) & (df['horsepower'] > 126) & (df['weight'] >=2800) & (df['weight'] <=3600)]
```

## Engineers at XYZ custom cars want to know how many cars are SUVs.

## Horsepower>140 , Weight>4500

```python
df.loc[(df['horsepower'] > 140) & (df['weight'] >=4500)]
```

## Engineers at XYZ custom cars want to how many cars are Racecars.

## Weight < 2223 , acceleration > 17

```python
df.loc[(df['acceleration'] > 17) & (df['weight'] < 2223)]
```

## XYZ custom cars want the data sorted according to the number of cylinders.

```python
df.sort_values(by = 'cylinders')
```

## There is a requirement in which the cars that have lowest acceleration must be assessed.It is also to be checked that whics cars have higher horsepower despite having lower acceleration.

```python
df.sort_values(['acceleration','horsepower'], ascending = (1,0))
```

## The board of XYZ custom cars wants to know about minimum and maximum sum, mean and median of all the numerical columns

```python
#Using List comprehension to get the numerical columns
list1 = [col for col in df.columns if df[col].dtype in ['float' , 'int64']]
df[list1].agg(['min' , 'max' , 'sum' , 'mean' , 'median'])
```

## XYZ custom cars want to know the number of cars manufactured in each year.

```python
df.groupby(['model_year']).count()['horsepower'] #All values will be same, so display any one column
```

## The engineers at XYZ custom cars want to know about the relationship between model year and acceleration of cars.

```python
df.groupby(['model_year']).mean()[['acceleration']]
```

## The engineers at XYZ custom cars wants to know the frequency distribution of different number of cylinders across different years.

```python
pd.crosstab(df['model_year'], df['cylinders'])
```

## The engineers at XYZ custon cars want to know the mean of all the numerical attributes of cars for each year

```python
pivot1 = pd.pivot_table(df, index = 'model_year' , aggfunc=np.mean)
pivot1
```
